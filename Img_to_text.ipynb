```python
from PIL import Image
import pytesseract
import cv2
import numpy as np
from pytesseract import Output
import re
```

```python
image = cv2.imread('sample.png')
```

```python
def get_grayscale(image):
  return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = get_grayscale(image)
Image.fromarray(gray)
```

```python
def remove_noise(image):
  return cv2.medianBlur(image, 5)
noise = remove_noise(gray)
Image.fromarray(gray)
```

```python
def thresholding(image):
  return cv2.threshold(image, 0, 224, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
thresh = thresholding(gray)
Image.fromarray(thresh)
```

```python
def erode(image):
  kernel = np.ones((5,5), np.uint8)
  return cv2.erode(image, kernel, iterations = 1)
erode = erode(gray)
Image.fromarray(erode)
```

```python
def opening(image):
  kernel = np.ones((5,5), np.uint8)
  return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
opening = opening(gray)
Image.fromarray(opening)
```

```python
def canny(image):
  return cv2.Canny(image, 100,200)
canny = canny(gray)
Image.fromarray(canny)
```

```python
def deskew(image):
  coords = np.column_stack(np.where(image > 0))
  angle = cv2.minAreaRect(coords)[-1]
  if angle < -45:
    angel = -(90 + angle)
  else:
    angle = -angle
  (h,w) = image.shape[:2]
  center = (w//2, h//2)
  M = cv2.getRotationMatrix2D(center, angle, 1.0)
  rotated = cv2.warpAffine(image, M, (w, h), flags = cv2.INTER_CUBIC, borderMode = cv2.BORDER_REPLICATE)
  return rotated
rotated = deskew(gray)
Image.fromarray(rotated)
```

```python
def match_template(image, template):
  return cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
match = match_template(gray, gray)
match
```

```python
img = cv2.imread('sample.png')
h, w, c = img.shape
boxes = pytesseract.image_to_boxes(img)
for b in boxes.splitlines():
  b = b.split(' ')
  x1, y1, x2, y2 = int(b[1]), int(b[2]), int(b[3]), int(b[4])
  img = cv2.rectangle(img, (x1, h - y2), (x2, h - y1), (0, 255, 0), 2)
Image.fromarray(img)
```

```python
img = cv2.imread('sample.png')
d = pytesseract.image_to_data(img, output_type = Output.DICT)
keys = list(d.keys())

data_pattern = 'text'

n_boxes = len(d['text'])
for i in range(n_boxes):
  if int(d['conf'][i]) > 60:
    if re.match(data_pattern, d['text'][i]):
      (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
      img = cv2.rectangle(img, (x,y), (x + w, y + h), (0, 225, 0), 2)
Image.fromarray(img)
```
